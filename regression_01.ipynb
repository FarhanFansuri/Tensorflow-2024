```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

#Permasalahan Harga Rumah

```python
# Contoh data untuk fitur dan target
# Fitur: luas tanah, jumlah kamar tidur
# Target: harga rumah
X = np.array([[1000, 3],
              [1500, 4],
              [1200, 3],
              [1800, 4]])

y = np.array([300000, 400000, 320000, 450000])

model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[2])])
model.compile(optimizer=tf.optimizers.Adam(), loss= 'mse', metrics=['acc'])
model.fit(X,y,epochs=100)
```
